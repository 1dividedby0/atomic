```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def S11(kappa1, kappa2, delta):
    return 1 - kappa1 / (1j * delta + (kappa1 + kappa2)/2)

def S12(kappa1, kappa2, delta):
    return -np.sqrt(kappa1 * kappa2) / (1j * delta + (kappa1 + kappa2)/2)

kappa_total = 1
delta = np.linspace(-3 * kappa_total, 3 * kappa_total, 100)
```

```python
kappa2 = 0

s11 = np.array([S11(kappa_total-kappa2, kappa2, i) for i in delta])
s12 = np.array([S12(kappa_total-kappa2, kappa2, i) for i in delta])

plt.plot(delta, np.abs(s11)**2, label=r'$|s_{11}|^2$')
plt.plot(delta, np.abs(s12)**2, label=r'$|s_{12}|^2$')
plt.plot(delta, np.angle(s11), label=r'Angle of $s_{11}$')
plt.plot(delta, np.angle(s12), label=r'Angle of $s_{12}$')
plt.title(r'$\kappa_2=0$')
plt.legend()
```

```python
kappa2 = kappa_total/2

s11 = np.array([S11(kappa_total-kappa2, kappa2, i) for i in delta])
s12 = np.array([S12(kappa_total-kappa2, kappa2, i) for i in delta])

plt.plot(delta, np.abs(s11)**2, label=r'$|s_{11}|^2$')
plt.plot(delta, np.abs(s12)**2, label=r'$|s_{12}|^2$')
plt.plot(delta, np.angle(s11), label=r'Angle of $s_{11}$')
plt.plot(delta, np.angle(s12), label=r'Angle of $s_{12}$')
plt.title(r'$\kappa_2=\kappa_1$')
plt.legend()
```

```python
kappa2 = 5*kappa_total/6

s11 = np.array([S11(kappa_total-kappa2, kappa2, i) for i in delta])
s12 = np.array([S12(kappa_total-kappa2, kappa2, i) for i in delta])

plt.plot(delta, np.abs(s11)**2, label=r'$|s_{11}|^2$')
plt.plot(delta, np.abs(s12)**2, label=r'$|s_{12}|^2$')
plt.plot(delta, np.angle(s11), label=r'Angle of $s_{11}$')
plt.plot(delta, np.angle(s12), label=r'Angle of $s_{12}$')
plt.title(r'$\kappa_2=5\kappa_1$')
plt.legend()
```

```python
delta = np.linspace(-30 * kappa_total, 30 * kappa_total, 10000)
```

```python
kappa2 = 0

s11 = np.array([S11(kappa_total-kappa2, kappa2, i) for i in delta])

plt.scatter(s11.real, s11.imag)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title(r'$\kappa_2=0$, Over-Coupled')
```

```python
kappa2 = kappa_total/2

s11 = np.array([S11(kappa_total-kappa2, kappa2, i) for i in delta])

plt.scatter(s11.real, s11.imag)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title(r'$\kappa_2=\kappa_1$, Critically Coupled')
```

```python
kappa2 = 5*kappa_total/6

s11 = np.array([S11(kappa_total-kappa2, kappa2, i) for i in delta])

plt.scatter(s11.real, s11.imag)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title(r'$\kappa_2=5\kappa_1$, Under-Coupled')
```
